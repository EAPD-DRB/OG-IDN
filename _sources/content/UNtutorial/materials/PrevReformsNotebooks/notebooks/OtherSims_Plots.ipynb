```python
def vat_revenue(tau_c, c, p_i, T=320):
    """
    Calculate VAT revenue
    """
    # Calculate population weights
    pop_weights = np.ones(T, 80, 7)/ (80*7)
    # Calculate VAT revenue
    cons_vat_revenue = (
        (
            ((tau_c[: T, :] * p_i).reshape(T, 2, 1, 1) * c).sum(
                axis=1
            )
            * pop_weights
        )
        .sum(1)
        .sum(1)
    )
    return cons_vat_revenue
```
